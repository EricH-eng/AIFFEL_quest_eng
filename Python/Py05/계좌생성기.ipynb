```python
import random # Q1 계좌번호 랜덤 생성 모듈 불러오기

class Account: # Q1 Account 클래스 생성
    bank_name ='SC은행' # Q1 은행이름 고정
    account_num = 0 # Q2 처음 계좌 수 지정
    
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.deposit_count = 0 # Q7 입금 횟수 기록
        self.deposit_log = [] # Q10 입금 내역 기록
        self.withdraw_log = [] # Q10 출금 내역 기록

        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)
        self.account_number = str(num1) + '-' + str(num2) + '-' + str(num3) # Q1 '-'와 합치기 위해 str로 처리

        Account.account_num += 1

    def get_account_num(self): # Q3 계좌수 출력 메서드 생성
        print(Account.account_num)

    def deposit(self): # Q4 입금 메서드 생성
        while True:
            amount = int(input('입금할 금액을 입력하세요.'))

            
            if amount < 1:
                print('입금은 최소 1원 이상만 가능합니다.')
            
            else:
                self.balance += amount
                self.deposit_count += 1 # Q7 입금 시마다 입금 횟수 증가
                self.deposit_log.append(amount) # Q10 입금 내역 기록

                if self.deposit_count % 5 == 0: # Q7 5번 입금 시 적용
                    self.balance += self.balance * 0.01 # Q7 1퍼센트 이자 더하기
                    self.balance = int(self.balance) # Q7 정수 변환
                                    
                break                 
    
    def withdraw(self): # Q5 출금 메서드 생성
        while True:
            amount = int(input('출금할 금액을 입력하세요.'))
        
            if self.balance <= amount:
                print('출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.')

            else:
                self.balance -= amount
                self.withdraw_log.appned(amount) # Q10 출금 내역 기록
                break

    def display_info(self): #Q6 청보 출력 메서스 생성
        balance_str = f"{self.balance:,}"
        print(f"은행이름: {Account.bank_name}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {balance_str}원")
    
    def deposit_history(self): # Q10
        print(self.deposit_log)

    def withdraw_history(self): # Q10
        print(self.withdraw_log)       
    
```

```python
a = Account("대한", 800000) # Q8 여러 객체 생성
b = Account("민국", 900000)
c = Account("만세", 1000000)

acount_list = [a, b, c]
```

```python
for account in acount_list: # Q9
    if account.balance >= 1000000:
        account.display_info()
```
